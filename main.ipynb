```python
import numpy as np
import torch
from torch import nn
import torch.nn.functional as F
from base_model import MyModel
import matplotlib.pyplot as plt
torch.set_grad_enabled(False)
```

```python
model = MyModel(input_size=3, hidden_sizes=[5, 4], output_size=1,
                weight_range=[-5, 5], bias_range=[-5, 5],
                weight_range_size=100, bias_range_size=100, num_examples=100)
model.eval()
model.evaluate()
```

```python
X = torch.randn(1, 3)
y = model(X)
print(f'True: {X.sum()}, Pred: {y}')

model.state_dict()
```

```python
history = model.search(100)
model.evaluate()
```

```python
plt.plot(history)
plt.xlabel('Iteration')
plt.ylabel('MSE Loss')
plt.title('Training History for Non Gradient Based Learning')
plt.show()
```
